```python
def convert_euro(euro):
    euro = euro.replace(' €','')
    euro = euro.replace(' ','')
    euro = int(euro)
    convert = euro*1.7361
    return round(convert,1)
# Qiymət sütunumuzda olan euro dəyərlərimizi azn valyutasına çevirmək üçün convert_euro() adlı funksiya təyin etdik
```

```python
def convert_usd(usd):
  usd = usd.replace(' $','')
  usd = usd.replace(' ','')
  usd = int(usd)
  convert = usd*1.70
  return round(convert,1)

# Qiymət sütunumuzda olan dollar dəyərlərimizi azn valyutasına çevirmək üçün convert_usd() adlı funksiya təyin etdik
```

```python
def replace_azn(azn):
    azn = azn.replace(' AZN','')
    azn = azn.replace(' ','')
    return azn
# Qiymət sütunumuzda olan AZN dəyərlərimizin AZN yazısını ixtisar etmək üçün  replace_azn() adlı funksiya təyin etdik
```

```python
data.info()
```

```python
# Qiymət sütunumuzda euro dəyərləri manatla əvz etmək üşün convert_euro() funksiyamızı pandasın apply() və pythonun lambda funksiyası vasitəsiylə datamıza tətbiq edirik
data['Qiymət'][data['Qiymət'].astype(str).str.contains(" €")]=data['Qiymət'][data['Qiymət'].astype(str).str.contains(" €")].apply(lambda x : convert_euro(x))
```

```python
# Qiymət sütunumuzda dollar dəyərləri manatla əvz etmək üşün convert_usd() funksiyamızı pandasın apply() və pythonun lambda funksiyası vasitəsiylə datamıza tətbiq edirik

data['Qiymət'][data['Qiymət'].astype(str).str.endswith(" $")] = data['Qiymət'][data['Qiymət'].astype(str).str.endswith(" $")].apply(lambda x : convert_usd(x))
```

```python
# Qiymət sütunumuzda AZN dəyərlərdəki azn sözünü ixtisar etmək üçün replace_azn() funksiyamızı pandasın apply() və pythonun lambda funksiyası vasitəsiylə datamıza tətbiq edirik
data['Qiymət'][data['Qiymət'].astype(str).str.endswith(" AZN")] = data['Qiymət'][data['Qiymət'].astype(str).str.endswith(" AZN")].apply(lambda x : replace_azn(x))
```

```python
# Outlayerlərimizin silinməsi
Q1 = np.percentile(data['Marka'], 25,
                   interpolation = 'midpoint')

Q3 = np.percentile(data['Marka'], 75,
                   interpolation = 'midpoint')

IQR = Q3 - Q1
upper = np.where(data['Marka'] >= (Q3+1.5*IQR))
lower = np.where(data['Marka'] <= (Q1-1.5*IQR))
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
```

```python
data.boxplot(column='Model');
```

```python
for x in ['Model']:
    q75,q25 = np.percentile(data.loc[:,x],[75,25])
    intr_qr = q75-q25

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)

    data.loc[data[x] < min,x] = np.nan
    data.loc[data[x] > max,x] = np.nan

# Outlayerlerin silinmesi ucun prosesi 3 defe tekrarlayiriq
```

```python
import optuna

def objective(trial):

  n_estimators =  trial.suggest_int('n_estimators', 30, 10000)
  criterion = trial.suggest_categorical('criterion', ['squared_error', 'absolute_error','poisson'])
  max_depth = trial.suggest_int('max_depth', 2, 150)
  min_samples_split = trial.suggest_int('min_samples_split', 2, 100)
  min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 100)
  #min_weight_fraction_leaf = trial.suggest_int('min_weight_fraction_leaf',0.0,20.0)
  max_features = trial.suggest_categorical('max_features', ['sqrt','log2'])
  #max_leaf_nodes = trial.suggest_int('max_leaf_nodes',None,10.0)
  #min_impurity_decrease = trial.suggest_int('min_impurity_decrease',0.0,20.0)
  max_samples = trial.suggest_int('max_samples', 1, 100)





  random_regression = RandomForestRegressor(n_estimators = n_estimators,
                                             criterion = criterion,
                                             max_depth = max_depth,
                                             min_samples_split = min_samples_split,
                                             min_samples_leaf = min_samples_leaf,
                                             #min_weight_fraction_leaf = min_weight_fraction_leaf,
                                             max_features = max_features,
                                             #max_leaf_nodes = max_leaf_nodes,
                                             #min_impurity_decrease = min_impurity_decrease,
                                             max_samples = max_samples,
                                             n_jobs=2)


  random_regression.fit(X_train, y_train)

  y_predict = random_regression.predict(X_test)


  return random_regression.score(X_test,y_test)
```

```python
random_study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.RandomSampler(),
)

tpe_study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.TPESampler(),
)

cmaes_study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.CmaEsSampler(),
)
```

```python
cmaes_study.optimize(objective, n_trials=50)
tpe_study.optimize(objective, n_trials=50)
random_study.optimize(objective, n_trials=50)
```

```python
print('CMAES with RandomForestClassifier: ',round(cmaes_study.best_value*100,2))
print('TPE with RandomForestClassifier: ',round(tpe_study.best_value*100,2))
print('Random study with RandomForestClassifier:', round(random_study.best_value*100,2))
```
